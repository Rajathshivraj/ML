```python
# importin matplot lib library
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

plotting the data

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
 #cosine waaves
 plt.plot(x,z)
 plt.show()
```

```python
# adding title x axis and y axis labels
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),"g-")
plt.plot(x,np.cos(x),"r--")
plt.show()
```

```python
#Bar plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.xlabel("lanagauages")
plt.ylabel("number of students")
ax.bar(langs,students)
plt.show()
```

pie chart

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students,labels=langs,autopct='%1.8f%%')
plt.show()
```

scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z=np.cos(x)
plt.scatter(x,y)
plt.show()
```

```python
x=np.linspace(10,40,100)
y=np.random.randint(10,40,100)
z=np.random.randint(10,40,100)
plt.scatter(x,y,c=z)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(10,40,100)
y=np.random.randint(10,40,100)
z=np.random.randint(10,40,100)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,y,c=z,cmap='viridis')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('scatter plot')
ax.scatter(x,y,color="g")
ax.scatter(x,y,marker='x')
ax.scatter(x,z,color='r')
plt.show()
```

3D SCATTER PLOT

```python
fig3 = plt.figure()
ax3 = plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax3.scatter3D(x,y,z,c=z,cmap='Blue')
plt.show()
```
